```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.Figure(figsize=(10,10))

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.random.uniform(0, 10, 50) 
y = 1.6*x+np.random.normal(0, 5, 50)  # random Linear relation with noise

plt.scatter(x,y)

plt.title("Scatterplot",fontsize=20,fontweight='extra bold',color='red')
plt.xlabel("x --->", fontsize=15,fontweight='bold')
plt.ylabel("y --->", fontsize=15,fontweight='bold')
```

# Visualization with Matplotlib 
Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib
import matplotlib.pyplot as plt      # .pyplot() is the most common function we gonna use 
```

### Plotting linear graph (line plot)

```python
'''ploting linear graph'''

x = [1,2,3,4,5]
y = [10,25,15,45,50]

plt.plot(x,y)
```

```python
'''Giving infos'''
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic  Graph')

print(plt.plot(x,y))
```

```python
'''Customizing the line plot'''
```

```python
# chaning the color
plt.plot(x,y,color = 'g')

```

```python
# To know how many colors are there
import matplotlib.colors


print(matplotlib.colors.BASE_COLORS)
print(matplotlib.colors.CSS4_COLORS)

# and many more are there
```

```python
'''design of line'''

plt.plot(x,y,linestyle = 'dotted')

# supported linestyle : '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
'''Change length of line'''

plt.plot(x,y,linewidth = 10)
```

```python
'''Marker Width'''

plt.plot(x,y,marker = 'o',markersize= '10')
```

```python
'''To get grid'''

plt.plot(x,y)
plt.grid(True)
```

```python
''' Figure function'''

# This is used to customize the size of graph
```

```python
# Syntax : plt.figure(figsize=(width,height))

plt.figure(figsize=(5,3))

plt.plot(x,y)
```

```python
# For customizing the resolution

plt.figure(figsize=(2,1),dpi=250)

plt.plot(x,y)

'''Note that : the size of graph on screen is product of resolution and figsize'''
```

```python
'''To change colour of background'''

plt.figure(figsize=(5,3),facecolor='green')
plt.plot(x,y)
```

```python
'''Understanding subplots'''

### syntax : plt.subplot(nrows,ncoloumns,position)
### Note : subplot command always go above of plot() command
x = [i for i in range(10)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.figure(figsize=(15,5),dpi=100)
# plt.grid(True)

#1 :
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Plot 1")

#2 :
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Plot 2")

#3:
plt.subplot(2,2,3)
plt.plot(y1,y2)
plt.title('Plot3')
```

### Bar Graph

```python
names = ["A","B","C","D"]
bodycount = [10,12,12,8]

plt.figure(figsize = (5,3),dpi=100,facecolor='purple')

plt.bar(names,bodycount,color = 'red')
plt.xlabel('Names')
plt.ylabel('Bodycount')
plt.title('Bargraph')
plt.show()
```

### Histograms
These used for data distribution
They distribute the data into bins and then plots the number of data elements in each bin


```python
data = [1,2,2,3,4,4,4,4,5,5,5,2,3]

plt.figure(figsize=(5,3))
plt.hist(data,bins = 4,color='pink',edgecolor = 'black')
```

### Pie charts

```python
names = ['A','B','C','D']     #The data that needs to be displayed - labels
count = [10,20,30,90]         #Mathematical data of given data
color = ['y','b','g','r']     #if not specified, systum will give random colors
explode = (0,0.2,0,0)         #will move that respective part given centimeter away

plt.pie(count,labels = names,colors=color,explode=explode,shadow=True,autopct='%.2f%%')

# "%.1f%%": Shows one decimal place, e.g., 25.3%
# "%.2f%%": Shows two decimal places, e.g., 25.34%
# "%d%%": Shows no decimal places, rounding to the nearest integer, e.g., 25%
```

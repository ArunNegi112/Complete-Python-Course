## Seaborn 

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns 
```

```python
tips = sns.load_dataset('tips')     # This function has some datasets 
tips.head()
```

```python
# Scatter plot

sns.scatterplot(data = tips, x='total_bill',y='tip')


# Notice that we dont have to use pandas and matplotlib both here 
```

```python
# Line plot 

sns.lineplot(data = tips,x='size',y='tip')

'''Note : The shown line is according to the avg data, and the spreaded region is called *CONFIDENCE INTERVAL* this show all the possible values given data can have'''
```

```python
# Line plot without confidence interval

sns.lineplot(data = tips,x='size',y='tip',ci=None)
```

```python
# Bar plot

sns.barplot(data = tips,x= 'size',y='tip')

# check for yourself
import pandas as pd
print(tips.groupby('size')['tip'].mean())
```

```python
# Box plot

sns.boxplot(data = tips, x = 'size', y = 'tip')
```

```python
# violin plot

sns.violinplot(data = tips, x = 'size',y = 'tip')
```

```python
# Histogram

sns.histplot(data = tips,x = 'size',bins=5)
```

```python
# Histogram with a curve 

sns.histplot(data = tips,x = 'size',bins=5,kde = True)
```

```python
# KDE plot ( Kernal Density Estimation)

sns.kdeplot(data = tips,x = 'size',fill= True)
```

```python
# Pairplot : This is used to differentiate betweeen graphs

sns.pairplot(data = tips)
```

```python
# Heatmap 

correlation = tips[['total_bill','tip','size']].corr()
correlation


# '''This mean, ex : when total bill goes up with 1, tip will go up with 0.6'''
```

```python


sns.heatmap(
    correlation,
    cmap='Blues',  #This is to change color type
    annot = True, #This to give those numbers on squares
    annot_kws={'size':15}
    )
 
```

```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
plt.figure(figsize=(5,3))

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.random.uniform(0, 10, 50) 
y = 1.6*x+np.random.normal(0, 5, 50)  # random Linear relation with noise

plt.scatter(x,y)

plt.title("Scatterplot",fontsize=20,fontweight='extra bold',color='red')
plt.xlabel("x --->", fontsize=15,fontweight='bold')
plt.ylabel("y --->", fontsize=15,fontweight='bold')
```

#### ___Customizing it___

```python
plt.figure(figsize=(5,3))


np.random.seed(42) 
x = np.random.uniform(0, 10, 50) 
y = 1.6*x+np.random.normal(0, 5, 50)



plt.scatter(
    x=x,
    y=y,
    color='r',          # list of colors can also be give 
    s=90,               # size of marker
    alpha=1,            # transparancy 
    marker='*',         # marker shape (can also give list)
    linewidth=0.5,      # linewidth of shapes
    edgecolor='b',      # color of edges of shapes
)


plt.title("Scatterplot",fontsize=20,fontweight='extra bold',color='red')
plt.xlabel("x --->", fontsize=15,fontweight='bold')
plt.ylabel("y --->", fontsize=15,fontweight='bold')
```

Note: To keep the same shade of colors, use colormap i.e cmap  
In cmap, each color is associated with a number, give list of numbers(1-100) to colors list.
then choose 'cmap' color theme in plt.scatter

```python
colors = [random.randint(1,100) for i in range(50)]
plt.scatter(x, y, c=colors, cmap='plasma')

plt.colorbar().set_label('colorbar',fontsize=15,fontweight='light',alpha=1)
```

```python
# How to use cmap
colors = x*y

plt.scatter(x,y,cmap='BrBG',c=colors)
plt.colorbar()
```

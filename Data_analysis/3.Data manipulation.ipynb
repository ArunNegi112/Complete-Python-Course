- Handling missing values

```python
csv_data.isnull()     
```

```python
csv_data.isnull().any()       # Tell if there is any missing value present in repective column
```

```python
csv_data.isnull().sum()    # Tell how many missing values are there
```

```python
csv_data.fillna(0)      # Fills the missing value with whatever we give here
```

```python
# Making a new column where all the missing values are filled with mean of that column

csv_data["Sales_filled"] = csv_data["Sales"].fillna(csv_data["Sales"].mean())
csv_data
```

```python
# Renaming Columns

csv_data = csv_data.rename(columns={"Sales_filled":"Sales_fillNA"})
csv_data.head()
```

```python
# Changing dtype of data

csv_data['Sales_fillNA_new'] = csv_data['Sales_fillNA'].astype(int)
csv_data.head()
```

```python
# Applying a function on data

csv_data['New_Value'] = csv_data['Value'].apply(lambda x:x+2)
csv_data.head()
```

### Data Agrregation and Grouping

data can be grouped by .groupby() function.
it identifies the same data in any row/column and group them together to get insights

```python
product_grouping = csv_data.groupby('Product') #Each unique value in the 'Product' column will form a separate group.

product_grouping
```

```python
product_average = csv_data.groupby('Product')['Value'].mean()
product_average
```

```python
Total_sales_in_every_region = csv_data.groupby(['Product','Region'])['Sales'].sum()
print(Total_sales_in_every_region)

# This is tell how much sale is n every region of every product

'''Note :
        This out put type is DataSeries'''
print(type(Total_sales_in_every_region))
```

```python
# Aggregating multliple functions

csv_data_insights = csv_data.groupby(['Product','Region'])['Sales'].agg(['sum','mean'])
csv_data_insights
```

```python
'''Merging different dataframes'''
df1 = pd.DataFrame({'key':['A',"B",'C','D'],'value1':[1,2,3,4]})
df2 = pd.DataFrame({'key':['A',"B",'C','E'],'value2':[10,20,30,40]})
```

```python
df1
```

```python
df2
```

```python
merged_df1 = pd.merge(df1,df2,how = 'inner')    # inner me common wale elements ate
merged_df1
```

```python
merged_df2 = pd.merge(df1,df2,how = 'outer')     # sare elements ate
merged_df2
```

```python
merged_df3 = pd.merge(df1,df2,how = 'left')   #left wala gets priority
merged_df3 
```

```python
merged_df4 = pd.merge(df1,df2,how = 'right')    # right wala gets priority
merged_df4
```
